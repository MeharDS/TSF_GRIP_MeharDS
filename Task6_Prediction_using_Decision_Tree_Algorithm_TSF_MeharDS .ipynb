# Model Building:

```python
final_df=df.drop(['species_encoded', 'Id'], axis=1)
```

```python
# Split data into X and y
X=final_df.drop('Species', axis=1)
y=final_df['Species']
```

## Create X (predictor ) and y (target) variable:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

## Now let us define the Decision Tree Algorithm:

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)

print('Decision Tree Classifer Created')
```

```python
y_pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy 
```

## Let us visualize the Decision Tree to understand it better:

```python
from sklearn.tree import plot_tree 
fig, ax = plt.subplots(figsize=(18,12))
plot_tree(dtree, feature_names=list(X.columns), class_names=list(final_df['Species'].unique()), filled=True)
plt.show();
```

### lets input a set of measurements of the iris flower and use the model to predict the iris species:

```python
x_new = np.array([[6, 3, 6, 2, 6]])
prediction = dtree.predict(x_new)
print("Prediction: {}".format(prediction))
```
