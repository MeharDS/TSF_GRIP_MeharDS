```python
final_df=df.drop('Species_encoded', axis=1)
```

## Create X (predictor ) and y (target) variable:

```python
X=final_df.drop('Species', axis=1)
y=final_df['Species']
```

## Split the data into test and train:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

## Model Building:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
knc= KNeighborsClassifier(n_neighbors=15,algorithm="auto")
rfc=RandomForestClassifier()
```

```python
#creating dictuory for clfs
clfs={
    'lr':LogisticRegression(),
    'knc':KNeighborsClassifier(),
    'rfc':RandomForestClassifier()
    }
```

```python
#Function for train model
def train_classifer(clf, X_train, X_test, y_train, y_test):
    clf.fit(X_train,y_train)
    print(f'{clf} Train Score is : ' , clf.score(X_train, y_train))
    print(f'{clf} Test Score is : ' , clf.score(X_test, y_test))
    y_pred = clf.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return score
```

```python
scores_list=[]
for name, clf in clfs.items():    
    current_score = train_classifer(clf, X_train, X_test, y_train, y_test)
    scores_list.append(current_score)
```

## Creating Performance Data Frame:

```python
performance_df = pd.DataFrame({'Algorithm': clfs.keys(), 'Score':scores_list})
```

```python
performance_df
```

```python
model=knc.fit(X_train, y_train)
y_pred = knc.predict(X_test)
```

```python
model.score(X_test, y_test)
```

## Print Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## let’s input a set of measurements of the iris flower and use the model to predict the iris species:

```python
x_new = np.array([[6, 3, 6, 2.5]])
prediction = model.predict(x_new)
print("Prediction: {}".format(prediction))
```
